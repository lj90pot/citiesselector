```python
# üîß Basic functions
def open_data(data): # returns shape, data types & shows a small sample
    print(f"Data shape is {data.info()}.\n")
    return data.sample(5)

def explore_data(data): # sum & returns duplicates, NaN & empty spaces added a table with percentage missing
    duplicate_rows = data.duplicated().sum()
    nan_values = data.isna().sum()
    empty_spaces = data.eq(' ').sum()
    exploration = pd.DataFrame({"NaN": nan_values, "EmptySpaces": empty_spaces}) # New dataframe with the results
    print(f"There are {data.duplicated().sum()} duplicate rows. Also;")
    nulls = pd.DataFrame(data.isna().sum()*100/len(data), columns=['percentage'])
    nulls.sort_values('percentage', ascending = False)
    print(nulls)
    return exploration
```

#### Join all tables to create a single database

A lot of nulls but this is in all years in next notebook the data is reduced to get the most updated for each city

Join all information

```python
#Function to join all tables with cities and time period
def table_join(tab1,tab2): 
    tab_final=pd.merge(left=tab1,right=tab2,how='outer',on=['cities','TIME_PERIOD'])
    tab_final=tab_final.reset_index(drop=True)
    return tab_final
```

```python
df_final=table_join(df_ceduc,df_cenv)
df_final=table_join(df_final,df_cfermor)
df_final=table_join(df_final,df_clivcon)
df_final=table_join(df_final,df_clma)
df_final=table_join(df_final,df_cpop1)
df_final=table_join(df_final,df_cpopcb)
df_final=table_join(df_final,df_ctran)
df_final=table_join(df_final,df_ctour)
df_final=table_join(df_final,df_cpopstr)
df_final=table_join(df_final,df_cecfi)
```
